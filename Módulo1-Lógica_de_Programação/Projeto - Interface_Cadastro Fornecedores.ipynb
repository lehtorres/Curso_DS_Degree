# Avaliação módulo 1 - Degree Data Science

Na empresa em que você trabalha foi solicitada a implementação de um sistema de cadastro dos fornecedores. As informações que devem ser cadastradas são:

- Código do fornecedor;
- Nome do fornecedor;
- Telefone do fornecedor;
- Email do fornecedor.

Para isso, você deverá criar um script para armazenar essas informações em alguma estrutura de dados.

Também como parte do sistema de cadastro, é necessário criar uma "interface de navegação" pelo sistema, que tenha as seguintes funcionalidades:

- Permite a exibição (usando o `print`) de todas as informaçoes de contatos específicos, a partir da busca pelo código do fornecedor;
- Permite a remoção e adição de fornecedores à estrutura de cadastro.

Requisitos:

- Requisito #1 - Não existe um limite para fornecedores cadastrados, portanto você deverá permitir que o usuário cadastre a quantidade que desejar.
- Requisito #2 - Crie pelo menos uma função em seu script. Você deverá escolher pelo menos uma tarefa do seu script e implementar em uma função para chamá-la dentro da sua lógica de cadastro.
- Requisito #3 - Crie pelo menos uma verificação em seu código, por exemplo: Se o dicionário estiver vazio, não permita a opção remover.

Tópicos avaliados:

- Variáveis;
- Tipos de dados;
- Operadores aritméticos;
- I/O de dados;
- Expressões lógicas e condicionais;
- Laços de repetição;
- Listas e dicionários;
- Funções.


```python
#FUNÇÕES A SEREM UTILIZADAS NO PROGRAMA

# Função para cadastro de novos fornecedores
def novo_cadastro (codigo, nome, telefone, email):
    
    cadastro_fornecedor["codigos"].append(codigo)
    cadastro_fornecedor["nomes"].append(nome)
    cadastro_fornecedor["telefones"].append(telefone)
    cadastro_fornecedor["emails"].append(email)
        
    return 

# Função para remoção de novos fornecedores
def remover_cadastro (codigo):
    
    idx_codigo = cadastro_fornecedor["codigos"].index(codigo)
    for k in cadastro_fornecedor:
        cadastro_fornecedor[k].remove(cadastro_fornecedor[k][idx_codigo])
    return 


#INÍCIO DA INTERFACE

# Criação do dicionário para armazenamento das informações de cadastro  

cadastro_fornecedor = {"codigos":[], "nomes":[], "telefones":[], "emails":[]}

# Inicializar variavel que será utilizada na interface com o usuário
opcao = 0

# Utilização de 'while' para garantir que a interface rode até que o usuário decida parar
while opcao != 4:
    
    opcao = int(input("Selecione a operação que deseja executar:\n1- Novo cadastro\n2- Remover cadastro\n3- Exibir cadastro\n4- Encerrar\n"))

# Se usuário escolher 1
 
    if opcao == 1:
        codigo = input("Insira o código do fornecedor a ser cadastrado: ")

# O sistema verifica se já não existe fornecedor e, se sim, avisa o usuário para não ter dados duplicados     
        if codigo in cadastro_fornecedor["codigos"]: 
            print("O fornecedor já possui cadastro!")

# se os dados não estiverem duplicados, o programa chama a função para cadastro novo        
        else:
            nome = input("Insira o nome do fornecedor: ")
            telefone = input("Insira o telefone do fornecedor: ")
            email = input("Insira o e-mail do fornecedor: ")
            novo_cadastro(codigo, nome, telefone, email)

# Verifica se o usuário quer continuar na interface
        prox = input("\nDeseja realizar nova operação (sim/nao)? ").lower()
        if prox == "não" or prox == "nao":
            opcao = 4 
        
# Se usuário escolher opção 2                      
    elif opcao == 2:
        codigo = input("Insira o código do fornecedor a ser removido: ")
        
# Se o código que o usuário deseja remover não está no dicionário, o usuário recebe um aviso, se estiver, chama a função para excluir os dados em todas as chaves
        if codigo not in cadastro_fornecedor["codigos"]:
            print("O fornecedor não está cadastrado ou já foi removido!")
        else:
            remover_cadastro(codigo)

# Verifica se o usuário quer continuar na interface
        prox = input("\nDeseja realizar nova operação (sim/nao)? ").lower()
        if prox == "não" or prox == "nao":
            opcao = 4 

# Se usuário escolher opção 3   
    elif opcao == 3:
        cod_busca = input("Digite o código do fornecedor que deseja exibir: ")

# Verifica se o código digitado está cadastrado e, se sim, imprime os valores relacionados aquele cadastro
        if cod_busca not in cadastro_fornecedor["codigos"]:
            print("Cadastro não encontrado!")
        else:
            idx_cod = cadastro_fornecedor["codigos"].index(cod_busca)
            dados_cod = []
            for v in cadastro_fornecedor.values():
                dados_cod.append(v[idx_cod])
            print("\nO código e as informações de nome, telefone e e-mail do contato procurado são, respectivamente: ", dados_cod)
        
# Verifica se o usuário quer continuar na interface       
        prox = input("\nDeseja realizar nova operação (sim/nao)? ").lower()
        if prox == "não" or prox == "nao":
            opcao = 4 

# Avisa caso a interface for encerrada
    elif opcao == 4:
        print("Interface encerrada!")

# Se o usuário digitar qualquer outro número, a interface avisa que não é uma opção e volta para o loop
    else:
        print("Desculpe, opção não encontrada!\n\n")
```
